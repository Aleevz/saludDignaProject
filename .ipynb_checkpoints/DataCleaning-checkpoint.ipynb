```python
from datetime import datetime, timedelta
import time
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set_theme()
```

```python
df_nuevo = pd.read_excel('data/ReporteTurnosPacientes.xlsx')
df_nuevo.head()
```

```python
df_nuevo["FechaID"] = df_nuevo["FechaID"].astype(str).str.zfill(8)
df_nuevo["TurnoHoraInicio"] = df_nuevo["TurnoHoraInicio"].astype(str)
df_nuevo["FechaHoraLLegada"] = pd.to_datetime(df_nuevo["FechaID"] + " " + df_nuevo["TurnoHoraInicio"], format="%Y%m%d %H:%M:%S")
df_nuevo["FechaHoraLLegada"] = pd.to_datetime(df_nuevo["FechaHoraLLegada"])
df_nuevo.head()
```

```python
df_nuevo.info()
```

```python
df_nuevo.drop(['PacienteID','TAPRecepcionCaja','PacienteSP','PacienteGenero','PacienteFechaNacimiento','PacienteCodigoPostal'], axis = 1, inplace = True)
```

```python
df_nuevo['TurnoMinutosEspera'] = df_nuevo.groupby('Sucursal')['TurnoMinutosEspera'].transform(lambda x: x.fillna(x.mean()))
df_nuevo['TAPRecepcionMinutos'] = df_nuevo['TAPRecepcionMinutos'].transform(lambda x: x.fillna(x.mean()))
```

```python
cajas_por_sucursal = {'SAN MARTIN TEXMELUCAN': 4}

df_nuevo['Cajas'] = df_nuevo['Sucursal'].map(cajas_por_sucursal)
```

```python
df_nuevo['Prioridad'] = np.random.choice([1, 0], size=len(df_nuevo), p=[0.6, 0.4])
df_nuevo.head()
```
