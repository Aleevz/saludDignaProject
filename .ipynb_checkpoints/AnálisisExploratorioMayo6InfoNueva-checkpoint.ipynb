## Diccionario de los 2 df

## 1.- Dataframe Turnos Pacientes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

turnos_pacientes = pd.read_excel('data/Turnos Pacientes.xlsx')
print(turnos_pacientes.shape)
turnos_pacientes.head()
```

```python
pd.isna(turnos_pacientes).sum()
```

## Nos interesan los generos de los pacientes, pero 1/4 parte del dataframe no cuenta con este valor y no se puede hacer inputación simple por media o mediana. Eliminamos el 25% de las filas?

```python
#turnos_pacientes = turnos_pacientes.fillna(df.mean())
turnos_pacientes['PacienteGenero'].unique()
```

```python
df = turnos_pacientes.dropna(subset=['PacienteGenero'])
print(df.shape)
pd.isna(df).sum()
```

### Faltan alrededor de 3215 datos en TAP Recepción Minutos que es la cantidad de minutos de tiempo de atención del paciente en recepción. Haré inputación simple por media porque siento que es lo mejor

```python
df['TAPRecepcionMinutos'] = df['TAPRecepcionMinutos'].fillna(df['TAPRecepcionMinutos'].mean())
```

```python
pd.isna(df).sum()
```

```python
df.describe()
```

- Máximo de tiempo que esperó el paciente desde que le asignaron turno hasta que fue llamado a recepción (TurnoMinutosEspera): 180 minutos o 3 hrs. Nada ideal pero realista, no deberíamos de quitarlo

- Máximo de TAPRecepcionMinutos (la cantidad de minutos de tiempo de atención del paciente en recepción): 250 minutos o 4 hrs. Me hace sentido porque es la suma de la variable anterior y el tiempo que tardo en la recepción, entonces siempre tiene que ser mayor que TAPMinutosEspera. Solo que no se si es normal que dures 1 hr en recepción, a menos que el sistema haya fallado y tuvieran que esperar y el tiempo siguiera corriendo.

- TAPRecepcionCaja es el id de la caja de recepcionista, ósea no importa.

```python
df['TiempoMinutosCaja'] = abs(df['TAPRecepcionMinutos']-df['TurnoMinutosEspera'])
print(df.shape)
pd.isna(df).sum()
```

```python
hoy = pd.Timestamp.today()
df['PacienteEdad'] = (hoy - df['PacienteFechaNacimiento']) // pd.Timedelta(days=365.25)
df['PacienteEdad'] = df['PacienteEdad'].astype(int)
```

```python
df['HoraLlegada'] = df['TurnoHoraInicio'].astype(str).str.slice(0, 2)
df['HoraLlegada'] = pd.to_numeric(df['HoraLlegada'], errors='coerce')
```

```python
df.describe()
```

```python
df.head(3)
```

### Hipótesis: Los adultos mayores tardan más en recepción porque son más tardados recabando su información

```python
def clasificar_edad(edad):
    if edad <= 30:
        return 'joven'
    elif edad <= 60:
        return 'adulto'
    else:
        return 'adulto_mayor'

df['Grupo'] = df['PacienteEdad'].apply(clasificar_edad)
df.head(1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.boxplot(data=df, x='Grupo', y='TiempoMinutosCaja', palette='Set2')

plt.title('Boxplot del tiempo en recepción por grupo de edad')
plt.xlabel('Grupo de edad')
plt.ylabel('Tiempo en recepción (minutos)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Eliminamos outliers. Eliminaremos todos los valores mayor a 40 ya que todos los datos parecen estar concentrados en los primeros 40 minutos.

```python
df=df[df['TiempoMinutosCaja'] < 40]
df.shape
```

```python
df.groupby('Grupo')['TiempoMinutosCaja'].describe()
```

En el grupo de adultos mayores el máximo es el menor de los tres pero el promedio es el mayor de los tres. Además, de que es el que menos valores tiene, lo que quiere decir que en promedio tardan más los adultos mayores en comparación a los otros dos grupos de personas. Pero la diferencia de tiempos es muy pequeña para decirse que es significante, solo son 12 segundos de diferencia. La hipotesis se rechaza.

```python
import seaborn as sns
import matplotlib.pyplot as plt

colores = ['skyblue', 'orange', 'salmon']

plt.figure(figsize=(15, 4))

for i, grupo in enumerate(df['Grupo'].unique()):
    plt.subplot(1, 3, i + 1)
    datos = df[df['Grupo'] == grupo]['TiempoMinutosCaja']
    plt.hist(datos, bins=10, color=colores[i], edgecolor='black')
    plt.title(f'{grupo.capitalize()}')
    plt.xlabel('Tiempo de recepción (min)')
    plt.ylabel('Frecuencia')
    plt.grid(True)

plt.tight_layout()
plt.suptitle('Histograma de tiempo de recepción por grupo de edad', fontsize=16, y=1.05)
plt.show()
```

```python
df.groupby('PacienteGenero')['TiempoMinutosCaja'].describe()
```

```python
df.groupby(['PacienteGenero', 'Grupo'])['TiempoMinutosCaja'].describe()
```

### Importante: En las 2 horas que más gente hay en el día que son a las 7 y 8 de la mañana. Son los que más tardan en estar en el mostrador pagando y recopilando sus datos. También en estas dos horas el tiempo que espera el paciente desde que le asignaron turno hasta que fue llamado a recepción es muchisimo mayor que con otras. Esperan 5 minutos en promedio cuando en los otros espean 2 minutos. Se debe de bajar esto.

```python
mean_table_genero = df.groupby(['HoraLlegada', 'PacienteGenero'])['TurnoMinutosEspera'].mean().unstack()
mean_table_genero
```

```python
mean_table_edades = df.groupby(['HoraLlegada','Grupo'])['TurnoMinutosEspera'].mean().unstack()
mean_table_edades
```

Por prioridades, los adultos mayores deben de pasar primero que todos los demás grupos. Las horas en las que si se cumple esto es: 6,7,8,14,15,18. Todas las demás horas son los últimos, se debería de cambiar eso.
