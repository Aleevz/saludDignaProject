```python
import numpy as np 
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv("muestras_clean.csv")
df["FechaHoraLLegada"] = pd.to_datetime(df["FechaHoraLLegada"])
df.head()
```

```python
df.info()
```

```python
# Crear el histograma
plt.figure(figsize=(14, 6))
sns.histplot(
    data=df,
    x="Hora",
    hue="Sucursal",
    multiple="dodge",
    bins=range(6, 20),
    palette="tab10"  # opcional, mejora visual
)

plt.title("Nmero de pacientes por hora del da y por sucursal", fontsize = 18)
plt.xlabel("Hora del da", fontsize = 14)
plt.ylabel("Cantidad de pacientes", fontsize = 14)
plt.xticks(range(6, 20))
plt.grid(axis='y', linestyle='--', alpha=0.5)

# No se necesita plt.legend() porque seaborn lo hace automticamente

plt.tight_layout()

plt.savefig("numpacientes.png", dpi=300)
```

```python
# Crear una copia con un poco de ruido para simular continuidad
import numpy as np
df["Hora_jitter"] = df["Hora"] + np.random.normal(0, 0.2, size=len(df))

sns.displot(
    data=df, x="Hora_jitter", hue="Sucursal", kind="kde", height=6, aspect=2, fill=True
)

plt.title("Densidad estimada de llegadas por hora del da y sucursal")
plt.xlabel("Hora del da")
plt.ylabel("Densidad")
plt.xlim(5.5, 20.5)
plt.grid(True, linestyle="--", alpha=0.3)
plt.show()
```

```python
colores = plt.get_cmap("tab10").colors
df["Sucursal"].value_counts().plot(kind="pie", autopct="%1.1f%%", figsize=(6, 6), colors=colores, title="Porcentaje de datos por sucursal")
plt.ylabel("")  # Oculta la etiqueta del eje Y
plt.tight_layout()
plt.savefig("%datos.png", dpi=300)
```

```python
import matplotlib.pyplot as plt

# Preparar los datos
sucursal_counts = df["Sucursal"].value_counts()
colores = plt.get_cmap("tab10").colors

# Crear la figura
fig, ax = plt.subplots(figsize=(5, 5))

# Pie sin etiquetas de slice (labels=[])
wedges, texts, autotexts = ax.pie(
    sucursal_counts,
    labels=[""] * len(sucursal_counts),  # oculta etiquetas
    autopct="%1.1f%%",
    colors=colores,
    startangle=90
)

# Agregar leyenda fuera del grfico
#ax.legend(wedges, sucursal_counts.index, title="Sucursal", loc="center left", bbox_to_anchor=(1, 0.5))
ax.set_title("Porcentaje de datos por sucursal")
plt.tight_layout()
plt.savefig("%datos.png", dpi=300)
```

```python
df = df[df['TurnoMinutosEspera'] < 100]
plt.figure(figsize=(10, 6))
sns.histplot(df["TurnoMinutosEspera"], bins=50, kde=True, color="steelblue")
plt.title("Distribucin de minutos de espera")
plt.xlabel("Minutos de espera")
plt.ylabel("Nmero de pacientes")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Sucursal", y="TurnoMinutosEspera", palette="tab10")
plt.title("Distribucin de minutos de espera por sucursal", fontsize = 18)
plt.ylabel("Minutos de espera", fontsize = 14)
plt.xlabel("Sucursal", fontsize = 14)
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.savefig('MinutosEsperaSuc.png', dpi=300)
```

# Datos tiempo

```python
df1 = pd.read_csv('data\DataTiempos.csv')
df1.head()
```

```python
df_grouped = df1.groupby('Sucursal')['TEPMinutos'].mean()
sucursales = ['COYOACAN', 'CULIACAN', 'CULIACAN\nCAADAS',
       'CULIACAN\nCOLEGIO\nMILITAR', 'CULIACAN\nLA CONQUISTA',
       'CULIACAN\nLAS TORRES', 'CULIACAN\nNAKAYAMA',
       'CULIACAN\nUNIVERSITARIOS']
```

```python
plt.figure(figsize = (12,8))
plt.bar(sucursales, df_grouped, color = 'skyblue')
plt.xlabel('\nSucursal', fontsize = 14)
plt.ylabel('Minutos de espera\n', fontsize = 14)
plt.title('Tiempo de espera promedio por sucursal', fontsize = 18)
plt.xticks(rotation = 0);
plt.savefig('TEPS.png', dpi = 300)
```

```python
df_grouped = df1.groupby('EstudioModalidad')['TEPMinutos'].mean()
estudios = ['Densitometria', 'Electrocardiograma', 'Laboratorio',
       'Mastografia', 'Nutricion', 'Optometria', 'Papanicolaou',
       'Rayos X', 'Resonancia\nMagnetica', 'Tomografia', 'Ultrasonido']
plt.figure(figsize = (12,8))
plt.barh(estudios, df_grouped, color = 'skyblue')
plt.xlabel('\nSucursal', fontsize = 14)
plt.ylabel('Minutos de espera\n', fontsize = 14)
plt.title('Tiempo de espera promedio por sucursal', fontsize = 18)
plt.xticks(rotation=0);
plt.savefig('TEPS.png', dpi = 300)
```

```python
import matplotlib.pyplot as plt

# Calcular el promedio y ordenar de mayor a menor
df1 = df1[df1['TEPMinutos'] < 240]
df_grouped = df1.groupby('EstudioModalidad')['TEPMinutos'].mean().sort_values(ascending=False)

# Usar el ndice ordenado como lista de estudios
estudios_ordenados = df_grouped.index.tolist()
valores = df_grouped.values

plt.figure(figsize=(12, 8))
plt.barh(estudios_ordenados, valores, color='skyblue')
plt.xlabel('\nMinutos de espera\n', fontsize=14)
plt.ylabel('\nEstudio', fontsize=14)
plt.title('Tiempo de espera promedio por estudio', fontsize=18)
plt.gca().invert_yaxis()  
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('TEPes.png', dpi=300)
plt.show()
```
