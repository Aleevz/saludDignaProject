## Diccionario de los 2 df

## 1.- Dataframe Turnos Pacientes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

turnos_pacientes = pd.read_excel('data/Turnos Pacientes.xlsx')
print(turnos_pacientes.shape)
turnos_pacientes.head()
```

```python
pd.isna(turnos_pacientes).sum()
```

## Nos interesan los generos de los pacientes, pero 1/4 parte del dataframe no cuenta con este valor y no se puede hacer inputacin simple por media o mediana. Eliminamos el 25% de las filas?

```python
#turnos_pacientes = turnos_pacientes.fillna(df.mean())
turnos_pacientes['PacienteGenero'].unique()
```

```python
df = turnos_pacientes.dropna(subset=['PacienteGenero'])
print(df.shape)
pd.isna(df).sum()
```

### Faltan alrededor de 3215 datos en TAP Recepcin Minutos que es la cantidad de minutos de tiempo de atencin del paciente en recepcin. Har inputacin simple por media porque siento que es lo mejor

```python
df['TAPRecepcionMinutos'] = df['TAPRecepcionMinutos'].fillna(df['TAPRecepcionMinutos'].mean())
```

```python
pd.isna(df).sum()
```

```python
df.describe()
```

- Mximo de tiempo que esper el paciente desde que le asignaron turno hasta que fue llamado a recepcin (TurnoMinutosEspera): 180 minutos o 3 hrs. Nada ideal pero realista, no deberamos de quitarlo

- Mximo de TAPRecepcionMinutos (la cantidad de minutos de tiempo de atencin del paciente en recepcin): 250 minutos o 4 hrs. Me hace sentido porque es la suma de la variable anterior y el tiempo que tardo en la recepcin, entonces siempre tiene que ser mayor que TAPMinutosEspera. Solo que no se si es normal que dures 1 hr en recepcin, a menos que el sistema haya fallado y tuvieran que esperar y el tiempo siguiera corriendo.

- TAPRecepcionCaja es el id de la caja de recepcionista, sea no importa.

```python
df['TiempoMinutosCaja'] = abs(df['TAPRecepcionMinutos']-df['TurnoMinutosEspera'])
print(df.shape)
pd.isna(df).sum()
```

```python
hoy = pd.Timestamp.today()
df['PacienteEdad'] = (hoy - df['PacienteFechaNacimiento']) // pd.Timedelta(days=365.25)
df['PacienteEdad'] = df['PacienteEdad'].astype(int)
```

```python
df['HoraLlegada'] = df['TurnoHoraInicio'].astype(str).str.slice(0, 2)
df['HoraLlegada'] = pd.to_numeric(df['HoraLlegada'], errors='coerce')
```

```python
df.describe()
```

```python
df.head(3)
```

### Hiptesis: Los adultos mayores tardan ms en recepcin porque son ms tardados recabando su informacin

```python
def clasificar_edad(edad):
    if edad <= 30:
        return 'joven'
    elif edad <= 60:
        return 'adulto'
    else:
        return 'adulto_mayor'

df['Grupo'] = df['PacienteEdad'].apply(clasificar_edad)
df.head(1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.boxplot(data=df, x='Grupo', y='TiempoMinutosCaja', palette='Set2')

plt.title('Boxplot del tiempo en recepcin por grupo de edad')
plt.xlabel('Grupo de edad')
plt.ylabel('Tiempo en recepcin (minutos)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Eliminamos outliers. Eliminaremos todos los valores mayor a 40 ya que todos los datos parecen estar concentrados en los primeros 40 minutos.

```python
df=df[df['TiempoMinutosCaja'] < 40]
df.shape
```

```python
df.groupby('Grupo')['TiempoMinutosCaja'].describe()
```

En el grupo de adultos mayores el mximo es el menor de los tres pero el promedio es el mayor de los tres. Adems, de que es el que menos valores tiene, lo que quiere decir que en promedio tardan ms los adultos mayores en comparacin a los otros dos grupos de personas. Pero la diferencia de tiempos es muy pequea para decirse que es significante, solo son 12 segundos de diferencia. La hipotesis se rechaza.

```python
import seaborn as sns
import matplotlib.pyplot as plt

colores = ['skyblue', 'orange', 'salmon']

plt.figure(figsize=(15, 4))

for i, grupo in enumerate(df['Grupo'].unique()):
    plt.subplot(1, 3, i + 1)
    datos = df[df['Grupo'] == grupo]['TiempoMinutosCaja']
    plt.hist(datos, bins=10, color=colores[i], edgecolor='black')
    plt.title(f'{grupo.capitalize()}')
    plt.xlabel('Tiempo de recepcin (min)')
    plt.ylabel('Frecuencia')
    plt.grid(True)

plt.tight_layout()
plt.suptitle('Histograma de tiempo de recepcin por grupo de edad', fontsize=16, y=1.05)
plt.show()
```
